```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargadel dataset
url = 'https://raw.githubusercontent.com/randres6/Coderhouse/main/Sheet_turismoreceptivo__.csv'
data = pd.read_csv(url)

data.head()
```

# Detección de valores nulos

Usando isnull sobre el dataset para mostrar el número de valores nulos en cada columna.

```python
valores_nulos = data.isnull().sum()
print(f"Valores nulos por columna:\n{valores_nulos}")

```

# Detección de valores duplicados

Usando duplicated() para indicar cuántas filas duplicadas existen en el dataset.

```python
duplicados = data.duplicated().sum()
print(f"Número de filas duplicadas: {duplicados}")
```

# Eliminar filas dupliacadas

Verificando en el dataset que existen filas duplicadas que efectivamente distorsionan el analisis, se procede a utilizar drop_duplicates() para modificar el dataset original y eliminar todas las filas duplicadas.

```python
data = data.drop_duplicates()

# Verificar que se han eliminado los duplicados
print(f"Dimensiones del dataset original: {data.shape}")
```

# Resumen de categorías en variables categóricas

El resumen de cada columna categórica nos muestra la frecuencia de cada categoría, ayudándonos a entender mejor la distribución de las categorías.

```python
# Listar las columnas categóricas
columnas_categoricas = data.select_dtypes(include=['object']).columns

# Mostrar un resumen de categorías por cada columna categórica
for col in columnas_categoricas:
    print(f"Resumen de la columna '{col}':")
    print(data[col].value_counts())
    print("\n")
```

Se usa columns para obtener un listado de todas las columnas del dataset

```python
columnas = data.columns
print(columnas)
```

# Incorporar columnas al dataset

Se calculan y añaden las columnas de AñoIngreso y Trimestre, puesto que servirá como dato para futuras visualizaciones.

```python
# Formato de fecha de la columna FechaIngreso
data['FechaIngreso'] = pd.to_datetime(data['FechaIngreso'], format='%d/%m/%Y', errors='coerce')

# Extraer el año y el trimestre
data['AñoIngreso'] = data['FechaIngreso'].dt.year.astype(int)
data['Trimestre'] = data['FechaIngreso'].dt.quarter.astype(int)


data.head()

```

# Grafico Univariado

Involucra una sola variable. Un gráfico univariado muestra la distribución de una variable, como un histograma o un diagrama de caja (box plot) que muestra solo una dimensión de datos.

```python
# Crea el count plot de los destinos
plt.figure(figsize=(14, 8))
sns.countplot(data=data, y='Destino', order=data['Destino'].value_counts().index, palette='viridis')

# Añade título y etiquetas
plt.title('Frecuencia de Destinos Visitados por los Turistas')
plt.xlabel('Cantidad de Turistas')
plt.ylabel('Destino')

# Mostrar la gráfica
plt.show()

```

```python
data['GastoTotal'] = pd.to_numeric(data['GastoTotal'], errors='coerce')

# Crea el box plot
plt.figure(figsize=(14, 8))
sns.boxplot(x='Transporte Internacional de Ingreso', y='GastoTotal', data=data, palette='viridis')

# Añade título y etiquetas
plt.title('Distribución del Gasto Total por Tipo de Transporte')
plt.xlabel('Tipo de Transporte')
plt.ylabel('Gasto Total')
plt.xticks(rotation=45)

# Muestra la gráfica
plt.show()



```

# Grafico Bivariado

Involucra dos variables. Un gráfico bivariado muestra la relación entre dos variables, como un diagrama de dispersión o un gráfico de línea con dos ejes.

```python
# Calcula el gasto promedio por nacionalidad
gasto_promedio_por_pais = data.groupby('Pais').apply(lambda x: (x['GastoTotal'] * x['Gente']).sum() / x['Gente'].sum())

# Grafica el gasto promedio por país
plt.figure(figsize=(14,8))
gasto_promedio_por_pais.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Gasto Promedio de los Turistas por Nacionalidad')
plt.xlabel('País')
plt.ylabel('Gasto Promedio')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()


```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['GastoAlojamiento'], data['GastoAlimentacion'], alpha=0.5, edgecolors='w', s=100)
plt.title('Relación entre Gasto en Alojamiento y Gasto en Alimentación')
plt.xlabel('Gasto en Alojamiento')
plt.ylabel('Gasto en Alimentación')
plt.grid(True)
plt.show()
```

# Grafico Multivariado

Involucra tres o más variables. Un gráfico multivariado muestra la relación entre múltiples variables, como en un gráfico de línea que desglosa una tercera variable con colores o diferentes líneas.

En este caso se muestra el calculo de cantidad de turistas agrupados por año de ingreso y donde cada línea representa un país a lo largo del tiempo que es el eje x. Se realiza un top10 ya que el grafico se satura al momento de mostrar toda la muestra, se ordena de mayor a menor.

```python
total_turistas = data.groupby(['AñoIngreso', 'Pais'])['Gente'].sum().reset_index()

# Filtra el top 10 países con mayor cantidad de turistas
top_10_paises = total_turistas.groupby('Pais')['Gente'].sum().nlargest(10).index
total_turistas_top10 = total_turistas[total_turistas['Pais'].isin(top_10_paises)]

# Ordena la leyenda según el top 10
ordered_top_10 = total_turistas_top10.groupby('Pais')['Gente'].sum().sort_values(ascending=False).index

# Crea la gráfica
plt.figure(figsize=(12, 8))
sns.lineplot(data=total_turistas_top10, x='AñoIngreso', y='Gente', hue='Pais', hue_order=ordered_top_10, marker='o')

# Ajusta las etiquetas del eje x
plt.xticks(total_turistas_top10['AñoIngreso'].unique().astype(int))

# Añade título plt.title('Cantidad Total de Turistas por País y Año (Top 10 Países)')
plt.xlabel('Año')
plt.ylabel('Cantidad de Turistas')

# Muestra la leyenda con el orden del top 10
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')

# Muestra la gráfica
plt.show()

```

```python
# Calcula cantidad de turistas por motivo de viaje y nacionalidad
motivo_viaje_por_pais = data.groupby(['Pais', 'Motivo'])['Gente'].sum().unstack().fillna(0)

# Grafica el motivo principal del viaje para los principales países
top_paises = motivo_viaje_por_pais.sum(axis=1).sort_values(ascending=False).head(10).index
motivo_viaje_top_paises = motivo_viaje_por_pais.loc[top_paises]

motivo_viaje_top_paises.plot(kind='bar', stacked=True, figsize=(14,8))
plt.title('Motivo Principal del Viaje por Nacionalidad (Top 10 Países)')
plt.xlabel('País')
plt.ylabel('Cantidad Total de Turistas')
plt.xticks(rotation=90)
plt.legend(title='Motivo de Viaje', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
